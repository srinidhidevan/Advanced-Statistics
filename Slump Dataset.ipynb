### SP, SLUMP & Strength contains OUTLIERS

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range    
```

```python
lrSP,urSP=remove_outlier(data['SP'])
data['SP']=np.where(data['SP']>urSP,urSP,data['SP'])
data['SP']=np.where(data['SP']<lrSP,lrSP,data['SP'])


lrSLUMP,urSLUMP=remove_outlier(data['SLUMP'])
data['SLUMP']=np.where(data['SLUMP']>urSLUMP,urSLUMP,data['SLUMP'])
data['SLUMP']=np.where(data['SLUMP']<lrSLUMP,lrSLUMP,data['SLUMP'])

lrStrength,urStrength=remove_outlier(data['Strength'])
data['Strength']=np.where(data['Strength']>urmiles,urmiles,data['Strength'])
data['Strength']=np.where(data['Strength']<lrmiles,lrmiles,data['Strength'])

data.shape
```
